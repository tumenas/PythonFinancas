```python
from bcb import IFDATA
import pandas as pd

# Origem dos dados: https://www3.bcb.gov.br/ifdata/
# Documentação da API: https://olinda.bcb.gov.br/olinda/servico/IFDATA/versao/v1/swagger-ui3#/

ifdata = IFDATA()

print("Endpoints disponíveis:")
ifdata.describe()
```

```python
# Lista de Relatórios Disponíveis

endpoint_relatorios = ifdata.get_endpoint('ListaDeRelatorio')
dados_relatorios = pd.DataFrame(endpoint_relatorios.get())
dados_relatorios
```

```python
# Cadastro das IFs (informacao por trimestre: AnoMes 202412, 202409, 202406, etc)

endpoint_cadastro = ifdata.get_endpoint('IfDataCadastro')
dados_cadastro = pd.DataFrame(endpoint_cadastro.get(AnoMes = 202412))
dados_cadastro
```
