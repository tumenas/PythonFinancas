```python
import pandas as pd

endereco_arquivo = r'C:\Users\FapesbAPP0043 2023\OneDrive\Felipe\3-Artigos\debentures\data\raw\Debentures.com.br_Caracteristica_em_25-10-2025_as_10-51-22.xls'

# Leitura do arquivo (a extensão é 'xls', mas o arquivo está formatado como csv separado por espaços)
dados = pd.read_csv(endereco_arquivo, sep='\t', header=2, encoding='cp1252')
```

```python
dados.head()
```

```python
# Converter a coluna para datetime (formato dia/mês/ano)
dados['Data de Emissao'] = pd.to_datetime(dados['Data de Emissao'], format='%d/%m/%Y', errors='coerce')

# Criar a coluna com o ano
dados['ano'] = dados['Data de Emissao'].dt.year

# Criar a coluna com o valor total emitido
dados['Quantidade Emitida'] = dados['Quantidade Emitida'].astype(float)
dados['Valor Nominal na Emissao'] = dados['Quantidade Emitida'].astype(float)
dados['valor_total'] = dados['Quantidade Emitida']*dados['Valor Nominal na Emissao']
```

```python
dados.columns

```

```python
tabela_emissoes_ano = pd.pivot_table(
    dados,
    index='ano',
    values='valor_total',
    aggfunc=['count', 'sum']
)

tabela_emissoes_ano
```

```python
tabela_emissoes_indexador = pd.pivot_table(
    dados,
    index='ano',
    columns = 'indice',
    values='valor_total',
    aggfunc=['count', 'sum']
).fillna(0).reset_index()

tabela_emissoes_indexador[tabela_emissoes_indexador['ano']>2000]
```

```python
tabela_emissoes_banco = pd.pivot_table(
    dados,
    index='Coordenador Lider',
    values='valor_total',
    aggfunc={'valor_total': ['count', 'sum']}  # usar dicionário
).fillna(0).reset_index()

# Achatar as colunas
tabela_emissoes_banco.columns = ['Coordenador Lider','count', 'sum']

tabela_emissoes_banco.sort_values('sum', ascending=False)
```

```python
# Criar coluna com a informação se a emissão é pela resolução ICVM 476
dados['cvm_476'] = [1 if item == 'DISPENSA ICVM 476/09' else 0 for item in dados['Registro CVM da Emissao']]
```

```python
tabela_emissoes_tipo = pd.pivot_table(
    dados,
    index='cvm_476',
    columns = 'Deb. Incent. (Lei 12.431)',
    values='valor_total',
    aggfunc= 'count' 
).fillna(0)

tabela_emissoes_tipo
```

```python
# Valor apenas das emissões depois de 2000

tabela_emissoes_tipo = pd.pivot_table(
    dados[dados['ano']>2000],
    index='cvm_476',
    columns = 'Deb. Incent. (Lei 12.431)',
    values='valor_total',
    aggfunc= 'sum' 
).fillna(0)

tabela_emissoes_tipo
```
