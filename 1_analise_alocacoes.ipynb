```python
import pandas as pd
import glob


endereco_arquivos = r"C:\\Users\\FapesbAPP0043 2023\\OneDrive\\Felipe\\3-Artigos\\debentures\\data\\raw\\Bloco 6"
lista_arquivos = glob.glob(endereco_arquivos + r'\*.csv')
```

```python
tabela_alocacao = pd.DataFrame()

for arquivo in lista_arquivos:
    dados = pd.read_csv(arquivo, encoding = 'cp1252', sep = ';')

    # Filtrar Debêntures
    dados = dados[dados['TP_ATIVO']=='Debênture simples']

    # Agrupar as alocacoes por emissor/dt_vencimento
    alocacao_fundos = (
    dados.groupby(['DT_COMPTC', 'EMISSOR', 'CPF_CNPJ_EMISSOR', 'DT_VENC'])['VL_MERC_POS_FINAL']
    .agg(['count', 'sum'])
    .reset_index())

    # Adicionar resultados na tabela 
    tabela_alocacao = pd.concat([tabela_alocacao, alocacao_fundos[alocacao_fundos['sum']>0]], ignore_index= True)
```
